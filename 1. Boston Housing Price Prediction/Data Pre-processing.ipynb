## Pre-Processing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
% matplotlib inline
```

```python
from sklearn.datasets import load_boston
boston_data = load_boston()
df = pd.DataFrame(boston_data.data, columns=boston_data.feature_names);
```

```python
X = df[['LSTAT']].values
y = boston_data.target
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X, y);
```

### Without Pre-Processing

```python
alpha = 0.0001
w_ = np.zeros(1 + X.shape[1])
cost_ = []
n_ = 100

for i in range(n_):
    y_pred = np.dot(X, w_[1:]) + w_[0]
    errors = (y - y_pred)
    w_[1:] += alpha * X.T.dot(errors)
    w_[0] += alpha * errors.sum()
    cost = (errors**2).sum() / 2.0
    cost_.append(cost)
```

```python
plt.figure(figsize=(8,6))
plt.plot(range(1, n_ + 1), cost_);
plt.ylabel('SSE')
plt.xlabel('Epoch');
```

### With Pre-Processing

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
X_std = sc_x.fit_transform(X)
y_std = sc_y.fit_transform(y.reshape(-1,1)).flatten()
```

```python
alpha = 0.0001
w_ = np.zeros(1 + X_std.shape[1])
cost_ = []
n_ = 100

for i in range(n_):
    y_pred = np.dot(X_std, w_[1:]) + w_[0]
    errors = (y - y_pred)
    w_[1:] += alpha * X_std.T.dot(errors)
    w_[0] += alpha * errors.sum()
    cost = (errors**2).sum() / 2.0
    cost_.append(cost)
```

```python
plt.figure(figsize=(8,6))
plt.plot(range(1, n_ + 1), cost_);
plt.ylabel('SSE')
plt.xlabel('Epoch');
```

#### MaxAbsScaler

#### Normalization
1. L1 normalization: Least Absolute Deviations the sum of absolute values is 1 in each row
2. L2 normalization: Least Squares ensure that the sum of squares is 1

```python
X = [[3.,4.,5.],[-2.,-3.,-4.],[0.,-5.,3.]] #fictitious data
```

```python
X_normalized = preprocessing.normalize(X, norm='l2')
X_normalized
```

```python
# Alternatively
normalizer = preprocessing.Normalizer().fit(X) # fit does nothing
normalizer
```

```python
normalizer.transform(X)
```

### Encoding categorical features

```python
source = ['australia', 'singapore', 'new zealand', 'hong kong']
label_enc = preprocessing.LabelEncoder()
src = label_enc.fit_transform(source)
```

```python
print("country to code mapping:\n")
for k, v in enumerate(label_enc.classes_):
    print(v, '\t', k)
```

```python
test_data = ['hong kong', 'singapore', 'australia', 'new zealand']
result = label_enc.transform(test_data)
print(result)
```
